# Compare energies in different methods for solar system

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Reading data

```python
E_Forward_Euler = np.loadtxt("H-FEuler.txt")
E_Sym_Euler = np.loadtxt("H-Sym_Euler.txt")
E_Heun = np.loadtxt("H-Heun.txt")
E_Stormer_Verlet = np.loadtxt("H-Stormer_Verlet.txt")
```

### Define time

```python
time = np.arange(1,365000)
```

### Plot Energies

```python
plt.figure(figsize = (20,10))
plt.plot(time,E_Forward_Euler,label = "Forward Euler",color = "red")
plt.plot(time,E_Sym_Euler, label = "Symplectic Euler", color = "green")
plt.plot(time,E_Heun, label = "Heun", color = "blue")
plt.plot(time,E_Stormer_Verlet, label = "Stormer_Verlet", color = "brown")
plt.xlim(0,365000)
plt.legend(loc = "upper left",fontsize = 18)
plt.savefig("Energy.png")
```
