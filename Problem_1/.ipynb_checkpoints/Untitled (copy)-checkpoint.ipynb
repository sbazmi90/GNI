## Forward Euler

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('data.txt')
```

```python
m = data[:,0]
xi = data[:,1]
yi = data[:,2]
zi = data[:,3]
vxi = data[:,4]
vyi = data[:,5]
vzi = data[:,6]
t_max = 1000
G = 2.95 * (10**-4)
delta =1
N = 2
u = np.zeros((1000,2),float)
k = np.zeros((1000,2),float)
```

```python
x = np.zeros((N,10),float)
y = np.zeros((N,10),float)
z = np.zeros((N,10),float)
dx =np.zeros((10,10),float)
dy = np.zeros((10,10),float)
dz = np.zeros((10,10),float)
#ux = np.zeros((N,10),float)
#uy = np.zeros((N,10),float)
#uz = np.zeros((N,10),float)
px = np.zeros((N,10),float)
py = np.zeros((N,10),float)
pz = np.zeros((N,10),float)
r = np.zeros((10,10),float)
ux = np.zeros((10),float)
uy = np.zeros((10),float)
uz = np.zeros((10),float)
```

```python
x.shape
```

```python
for i in range(10):
    x[0,i] = xi[i]
    y[0,i] = yi[i]
    z[0,i] = zi[i]
    px[0,i] = m[i]*vxi[i]
    py[0,i] = m[i]*vyi[i]
    pz[0,i] = m[i]*vzi[i]
```

```python
x[0,1]-x[0,2]
```

```python
for t in range(1,10000):
    for k in range(10):
        ux[k] = uy[k] = uz[k] = 0
        for i in range(10):
            for j in range(10):
                if(i != j):
                    dx[i,j] = x[t-1,i]-x[t-1,j]
                    dy[i,j] = y[t-1,i]-y[t-1,j]
                    dz[i,j] = z[t-1,i]-z[t-1,j]
                    r[i,j] = np.sqrt(dx[i,j]**2 + dy[i,j]**2 + dz[i,j]**2)
                    ux[i] = ux[i]-G*m[i]*m[j]*dx[i,j]/((r[i,j]+0.00000001)**3)
                    uy[i] = uy[i]-G*m[i]*m[j]*dy[i,j]/((r[i,j]+0.00000001)**3)
                    uz[i] = uz[i]-G*m[i]*m[j]*dz[i,j]/((r[i,j]+0.00000001)**3)
        for l in range(10):
            x[t,l] = x[t-1,l] + delta*px[t-1,l]/m[l]
            y[t,l] = y[t-1,l] + delta*py[t-1,l]/m[l]
            z[t,l] = z[t-1,l] + delta*pz[t-1,l]/m[l]
            px[t,l] = px[t-1,l] + delta*ux[l]
            py[t,l] = py[t-1,l] + delta*uy[l]
            pz[t,l] = pz[t-1,l] + delta*uz[l]
```

```python
for t in range(t_max):
    for i in range(10):
        for j in range(10):
            if (i == j):
                print('')
            else:
                dx = x[i] - x[j]
                dy = y[i] - y[j]
                dz = z[i] - z[j]
                r = np.sqrt(dx**2 + dy**2 + dz**2)
                r3 = r**3
                u[i] = u[i] - G*m[i]*m[j]/r
                px[i,j] = px[i,j] - (G*m[i]*m[j]*(x[i]-x[j]))/(r3)
                py[i,j] = py[i,j] - (G*m[i]*m[j]*(y[i]-y[j]))/(r3)
                pz[i,j] = pz[i,j] - (G*m[i]*m[j]*(z[i]-z[j]))/(r3)
        k[i] = k[i] + 0.5*(px[i,j]**2 + px[i,j]**2 + pz[i,j]**2)
    for i in range(10):
        x[i] = x[i] + delta*px[i]/m[i]
        y[i] = y[i] + delta*py[i]/m[i]
        z[i] = z[i] + delta*pz[i]/m[i]
        px[i] = px[i] + delta*px[i]/m[i]
        py[i] = px[i] + delta*px[i]/m[i]
        pz[i] = px[i] + delta*px[i]/m[i]
```
