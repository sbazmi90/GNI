# Stormer Verlet for solar system

### import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('data.txt')
```

```python
m = data[:,0]
xi = data[:,1]
yi = data[:,2]
zi = data[:,3]
vxi = data[:,4]
vyi = data[:,5]
vzi = data[:,6]
t_max = 1000
G = 2.95 * (10**-4)
delta =1
N = 365000
UE = np.zeros((N-1),float)
KE = np.zeros((N-1),float)
```

```python
x = np.zeros((N,10),float)
y = np.zeros((N,10),float)
z = np.zeros((N,10),float)
dx =np.zeros((10,10),float)
dy = np.zeros((10,10),float)
dz = np.zeros((10,10),float)
#ux = np.zeros((N,10),float)
#uy = np.zeros((N,10),float)
#uz = np.zeros((N,10),float)
px = np.zeros((N,10),float)
py = np.zeros((N,10),float)
pz = np.zeros((N,10),float)
r = np.zeros((10,10),float)
ux = np.zeros((10),float)
uy = np.zeros((10),float)
uz = np.zeros((10),float)
ep = 1e-6
```

```python
for i in range(10):
    x[0,i] = xi[i]
    y[0,i] = yi[i]
    z[0,i] = zi[i]
    px[0,i] = m[i]*vxi[i]
    py[0,i] = m[i]*vyi[i]
    pz[0,i] = m[i]*vzi[i]
```

```python
for t in range(1,N):
    for k in range(10):
        ux[k] = uy[k] = uz[k] = UE[k] = KE[k] = 0
    for i in range(10):
        KE[t-1] = KE[t-1] + 0.5 * (px[t-1,i] * px[t-1,i] + py[t-1,i] * py[t-1,i] + pz[t-1,i] * pz[t-1,i])
        for j in range(10):
            if(i != j):
                dx[i,j] = x[t-1,i]-x[t-1,j]
                dy[i,j] = y[t-1,i]-y[t-1,j]
                dz[i,j] = z[t-1,i]-z[t-1,j]
                r[i,j] = np.sqrt(dx[i,j]**2 + dy[i,j]**2 + dz[i,j]**2)
                UE[t-1]=UE[t-1] - G* m[i]*m[j] / r[i,j]
    for i in range(10):
        for j in range(10):
            if(i != j):
                dx[i,j] = x[t-1,i]-x[t-1,j]
                dy[i,j] = y[t-1,i]-y[t-1,j]
                dz[i,j] = z[t-1,i]-z[t-1,j]
                r[i,j] = np.sqrt(dx[i,j]**2 + dy[i,j]**2 + dz[i,j]**2)
                ux[i] = ux[i]-G*m[i]*m[j]*dx[i,j]/((r[i,j]+ep)**3)
                uy[i] = uy[i]-G*m[i]*m[j]*dy[i,j]/((r[i,j]+ep)**3)
                uz[i] = uz[i]-G*m[i]*m[j]*dz[i,j]/((r[i,j]+ep)**3)
    for l in range(10):
            px[t,l] = px[t-1,l] + 0.5*delta*ux[l]
            py[t,l] = py[t-1,l] + 0.5*delta*uy[l]
            pz[t,l] = pz[t-1,l] + 0.5*delta*uz[l]
    for l in range(10):
            x[t,l] = x[t-1,l] + delta*px[t,l]/m[l]
            y[t,l] = y[t-1,l] + delta*py[t,l]/m[l]
            z[t,l] = z[t-1,l] + delta*pz[t,l]/m[l]
    for k in range(10):
        ux[k] = uy[k] = uz[k] = 0
    for i in range(10):
        for j in range(10):
            if(i != j):
                dx[i,j] = x[t,i]-x[t,j]
                dy[i,j] = y[t,i]-y[t,j]
                dz[i,j] = z[t,i]-z[t,j]
                r[i,j] = np.sqrt(dx[i,j]**2 + dy[i,j]**2 + dz[i,j]**2)
                ux[i] = ux[i]-G*m[i]*m[j]*dx[i,j]/((r[i,j]+ep)**3)
                uy[i] = uy[i]-G*m[i]*m[j]*dy[i,j]/((r[i,j]+ep)**3)
                uz[i] = uz[i]-G*m[i]*m[j]*dz[i,j]/((r[i,j]+ep)**3)
    for l in range(10):
        px[t,l] = px[t,l] + 0.5*delta*ux[l]
        py[t,l] = py[t,l] + 0.5*delta*uy[l]
        pz[t,l] = pz[t,l] + 0.5*delta*uz[l]
```

```python
xsun = x[:,0];ysun = y[:,0];zsun = z[:,0];
xmer = x[:,1];ymer = y[:,1];zmer = z[:,1];
xven = x[:,2];yven = y[:,2];zven = z[:,2];
xear = x[:,3];year = y[:,3];zear = z[:,3];
xmar = x[:,4];ymar = y[:,4];zmar = z[:,4];
xjup = x[:,5];yjup = y[:,5];zjup = z[:,5];
xsat = x[:,6];ysat = y[:,6];zsat = z[:,6];
xura = x[:,7];yura = y[:,7];zura = z[:,7];  
xnep = x[:,8];ynep = y[:,8];znep = z[:,8];                                     
xplo = x[:,9];yplo = y[:,9];zplo = z[:,9];
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure(figsize=(20,10))
ax = plt.axes(projection='3d')
ax.plot3D(xsun,ysun,zsun,'red')
ax.plot3D(xmer,ymer,zmer,'black')
ax.plot3D(xven,yven,zven,'')
ax.plot3D(xear,year,zear,'blue')
ax.plot3D(xmar,ymar,zmar,'brown')
ax.plot3D(xjup,yjup,zjup,'green')
ax.plot3D(xsat,ysat,zsat,'violet')
ax.plot3D(xura,yura,zura,'pink')
ax.plot3D(xnep,ynep,znep,'orange')
ax.plot3D(xplo,yplo,zplo,'gray')
```

```python
H = np.zeros((N-1),float)
```

```python
for i in range(0,N-1):
    H[i] = UE[i] + KE[i]
```
