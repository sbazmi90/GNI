```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
O1 = np.loadtxt("Object1.txt")
O2 = np.loadtxt("Object2.txt")
O3 = np.loadtxt("Object3.txt")
O4 = np.loadtxt("Object4.txt")
O5 = np.loadtxt("Object5.txt")
O6 = np.loadtxt("Object6.txt")
O7 = np.loadtxt("Object7.txt")
O8 = np.loadtxt("Object8.txt")
O9 = np.loadtxt("Object9.txt")
H = np.loadtxt("Energy.txt")
```

```python
o1x = O1[:,1]
o1y = O1[:,2]

o2x = O2[:,1]
o2y = O2[:,2]

o3x = O3[:,1]
o3y = O3[:,2]

o4x = O4[:,1]
o4y = O4[:,2]

o5x = O5[:,1]
o5y = O5[:,2]

o6x = O6[:,1]
o6y = O6[:,2]

o7x = O7[:,1]
o7y = O7[:,2]

o8x = O8[:,1]
o8y = O8[:,2]

o9x = O9[:,1]
o9y = O9[:,2]

t = H[:,0]
px = H[:,1]
py = H[:,2]
lz = H[:,3]
E = H[:,4]
```
